### Tratamento dos Dados

#### Remove pessoas com data de nascimento menor que 1940

```python
df = df[df['Year_Birth'] >= 1940]
df[df['Year_Birth'] < 1940]['Year_Birth'].head(50) #Teste
```

#### Tratamento de nulos da coluna "renda"(Income)

```python
mediana_renda = df['Income'].median()
df['Income'] = df['Income'].fillna(mediana_renda)
df['Income'].isna().sum()
```

#### Tratamento da coluna "estado civil" (Marital_Status)

##### Averiguação de valores distintos

```python
df['Marital_Status'].unique()
```

```python
estado_civil_dic = {
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Single'
}

df['Marital_Status'] = df['Marital_Status'].replace(estado_civil_dic)

df['Marital_Status'].unique()
```

##### Transformação de data (string) para data real

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

#### Engenharia de Atributos (Gerar valor para análises)

##### Criação de coluna idade

```python
df['Idade'] = datetime.now().year - df['Year_Birth']
df['Idade'].head(50)
```

##### Criação da coluna "Gastos Totais"

```python
df['Gasto_Total'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts']
df['Gasto_Total'].head()
```

##### Criação da coluna "Total_Filhos"

```python
df['Total_Filhos'] = df['Kidhome'] + df['Teenhome']
df['Total_Filhos'].head()
```

##### Criação da coluna "Anos_De_Loja" 

```python
df['Anos_De_Loja'] = datetime.now().year - df['Dt_Customer'].dt.year
df.columns
```

##### Drop de colunas inuteis

```python
df = df.drop(columns=['Year_Birth', 'Kidhome', 'Teenhome'], errors='ignore')
df.head(50)
```

```python
df.to_csv('limpo_LCPA.csv', index=False)
```
